```python
# we seen about most popular starting point for the uber drives

starting_point=data["START*"].dropna()
#to make data frame value count
df=pd.DataFrame(starting_point.value_counts())
# to make data in ascending order so we have use this
df.sort_values(["START*"],ascending=False)
# make index in proper values 
df=df.reset_index()
# to make rename for columns and its start as count and index has starting_destination
df=df.rename(columns={'index':'start_destination','START*':'count'})
# find out the max value in count columns
df.loc[df['count']==max(df['count'])]
```

```python
df.loc[df['count']==min(df['count'])]
```

```python
# to find how many most popular dropping point for uber drives

stoping_point=data["STOP*"].dropna()
df=pd.DataFrame(stoping_point.value_counts())
df.sort_values(["STOP*"],ascending=False)

df=df.reset_index()
df=df.rename(columns={'index':'stoping_point','STOP*':'count'})
df.loc[df['count']==max(df['count'])]
```

```python
df
df.loc[df['count']==min(df['count'])]
```

```python
# take the mostly frequently used route by ubers

df=data.dropna()

# making dataframe to group the start and stop columns

df=pd.DataFrame(df.groupby(['START*','STOP*']).size())

# rename the columns the count to 0

df=df.rename(columns={0:'count'})

# make value in the order 

df=df.sort_values(['count'],ascending=False)

# find out max value in data

df.loc[df['count']==max(df['count'])]
```
